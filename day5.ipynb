```python
#condition statement(if)
age=30
if age>=30:
    print("you can vote")
else:
    print("you cannot vote")
```

```python
#more than 1 condition
marks=75
if marks>=90:
    print("grade a")
elif marks>=70:
    print("grade b")
elif marks>=50:
    print("grade c")
else:
    print("fail")



```

```python
#nested if
age=18
has_licence='yes'
if age>18:
    if has_licence=='yes':
            print("you can drive")
    else:
        print("go and take licence")
else:
     print("you are too young drive")
     
```

```python
#same if 2 condition satisfied
mark=55
attendance=76
if mark>55 and attendance>70:
    print("allowed for exam")
else:
    print("not allowed")
```

```python
#same if 2 condition satisfied(or)
mark=55
attendance=76
if mark>55 or attendance>70:
    print("allowed for exam")
else:
    print("not allowed")

```
